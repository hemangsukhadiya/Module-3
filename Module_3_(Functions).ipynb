1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def function(l):
  sum1=sum(filter(lambda val: val % 2 == 0, l))
  return sum1

n = int(input("Enter the number of element you want in list: "))
l=[]
for i in range(n):
  a=int(input("Enter the elements of list: "))
  l.append(a)
print("The sum of all the even numbers given in the list ",list(filter(lambda val: val % 2 == 0, l))," is",function(l))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
from math import sqrt
def Prm(number, itr):
    if itr == 1 or itr == 2:
        return True
    if number % itr == 0:
        return False
    if Prm(number, itr - 1) == False:
        return False
    return True

num = int(input("Enter the number you want to check prime or not: "))
itr = int(sqrt(num) + 1)
if Prm(num, itr) :
  print("Entered number is prime")
else:
  print("Entered number is not prime")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def pwr_gen(base, exponent):
    result = 1
    for i in range(exponent + 1):
        yield result
        result *= base

base = int(input("Input the base number: "))
exponent = int(input("Input the exponent: "))

power_gen = pwr_gen(base, exponent)
print(f"Powers of {base} up to exponent {exponent}:")
for power in power_gen:
    print(power)
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
subject_marks = [('English', 88), ('Science', 90), ('Maths', 97), ('Social sciences', 82)]
print("Original list of tuples:")
print(subject_marks)
subject_marks.sort(key=lambda x: x[1])
print("\nSorting the List of Tuples:")
print(subject_marks)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowel(string):
    vowels = ['a','e','i','o','u']
    return string not in vowels
string=list(input("Enter the string you want to remove vowel from: "))
b=filter(remove_vowel,string)
s=''.join(b)
print("String after removing vowels from it: ",s)
```
